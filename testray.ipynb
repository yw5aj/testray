```python
import trimesh
import numpy as np
```

```python
mesh = trimesh.load_mesh('test.stl')

ray_origins = np.array([[ 0.12801793, 24.5030052 , -5.        ],
                        [ 0.12801793, 24.5030052 , -5.        ]])
ray_directions = np.array([[-0.13590759, -0.98042506,  0.        ],
                           [ 0.13590759,  0.98042506, -0.        ]])
```

```python
ray_visualize = trimesh.load_path(np.hstack((
                                            ray_origins,
                                            ray_origins + 10 * ray_directions)).reshape(-1, 2, 3))
scene = trimesh.Scene([mesh, ray_visualize])
scene.show()
```

```python
locations, index_ray, index_tri = mesh.ray.intersects_location(
    ray_origins=ray_origins, ray_directions=ray_directions)
locations
```
